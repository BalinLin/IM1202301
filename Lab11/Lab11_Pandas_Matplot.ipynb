# Pandas

```python
import pandas as pd
df = pd.read_csv('got_imdb.csv')
print(df)
```

```python
print(df.iloc[0:4,0:2])
print("-" * 100) 
print(df.loc[0])
print("-" * 100) 
print(df.loc[0]['Title'])
```

```python
print(df.shape) # 回傳列數與欄數  
print("-" * 100)  
print(df.describe()) # 回傳描述性統計  
print("-" * 100)    
print(df.head(3)) # 回傳前三筆觀測值  
print("-" * 100)   
print(df.tail(3)) # 回傳後三筆觀測值  
print("-" * 100) 
print(df.columns) # 回傳欄位名稱  
print("-" * 100)  
print(df.index) # 回傳 index  
print("-" * 100)  
print(df.info) # 回傳資料內容  
```

# Matplot

```python
import matplotlib.pyplot as plt
x = df['AirDate']
y1 = df['Rating']
y2 = df['Season']
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y1, color=(255/255, 100/255, 100/255), marker = "o")
plt.plot(x, y2, '--', color=(0/255, 0/255, 255/255))
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(x, y1, color=(255/255, 100/255, 100/255), marker = "o")
plt.plot(x, y2, '--', color=(0/255, 0/255, 255/255))
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(x, y1, color=(255/255, 100/255, 100/255), marker = "o", label = 'Rate')
plt.plot(x, y2, '--', color=(0/255, 0/255, 255/255), label = 'Season')
plt.legend(loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(x, y1, color=(255/255, 100/255, 100/255), marker = "o", label = 'Rate')
plt.plot(x, y2, '--', color=(0/255, 0/255, 255/255), label = 'Season')
plt.legend(loc = 'upper right')
plt.title("Game of Thrones")
plt.ylabel("Value")
plt.xlabel("Date")
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(x, y1, color=(255/255, 100/255, 100/255), marker = "o", label = 'Rate')
plt.plot(x, y2, '--', color=(0/255, 0/255, 255/255), label = 'Season')
plt.legend(loc = 'upper right')
plt.title("Game of Thrones")
plt.ylabel("Value")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(x, y1, color=(255/255, 100/255, 100/255), marker = "o", label = 'Rate')
plt.plot(x, y2, '--', color=(0/255, 0/255, 255/255), label = 'Season')
plt.legend(loc = 'upper right')
plt.title("Game of Thrones")
plt.ylabel("Value")
plt.xlabel("Date")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
